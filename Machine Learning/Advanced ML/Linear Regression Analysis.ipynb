```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Read the Data From CSV File
df = pd.read_csv("homes.csv")
df
```

# Regression Between Sell and Living

```python
# Sell Data 
y = np.array(df.iloc[:,[0]])
Y = y.reshape(50)
Y
```

```python
# Living Data 
X = np.array(df.iloc[:,[2]])
X
```

```python
# Creating a Linear Model
model = linear_model.LinearRegression()
model.fit(X,Y)
```

```python
# Finding the Coefficent
model.coef_
```

```python
# Finding the intercept point
model.intercept_
```

```python
# Prediction on Training data
Yp = model.predict(X)
Yp
```

```python
# ploting in graph 
plt.plot(X,Yp,color='red')
plt.scatter(X,Y)
plt.xlabel("Living")
plt.ylabel("Sell")
plt.title("Sell VS Living Data Distribution along with the Regression Line")
plt.show()
```

```python
# Finding the correlations
corr = df.corr()
corr
```

```python
# Model Score
model.score(X,Y)
```
