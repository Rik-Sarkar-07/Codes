```python
import numpy as np
import matplotlib.pyplot as plt
```

# OR Gate Data

```python
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
y = np.array([0,1,1,1])
```

```python
plt.scatter(x1,x2, c=y, cmap='viridis', s=10)
plt.show()
```

```python
def Activation(x):
    if x >= 0.5:
        return 1 
    else:
        return 0

def preActivation(x1,x2,w1,w2):
    return float(w1*x1 + w2*x2)


def activationFunction(x1,x2,w1,w2):
    if preActivation(x1,x2,w1,w2) >= 0.5:
        return 1
    else:
        return 0

def errorFunction(x1,x2,w1,w2,y):
    return (y - preActivation(x1,x2,w1,w2))


def lossFunction(x1,x2,w1,w2,y):
    err = 0
    for i in range(len(y)):
        err += abs(y[i] - Activation((x1[i]*w1 + w2*x2[i])))
    return err
    
```

```python
def perceptronTraing(x1,x2,y):
    epoch = 0
    w1 = 0.1
    w2 = 0.2
    lemda = 0.1

    while (True):
        for i in range(len(y)):
            w1 += lemda*((y[i] - activationFunction(x1[i],x2[i],w1,w2))*x1[i])
            w2 += lemda*((y[i] - activationFunction(x1[i],x2[i],w1,w2))*x2[i])
        epoch += 1
       # print(epoch)
        if(lossFunction(x1,x2,w1,w2,y) == 0 or epoch == 100000):
            print('After '+str(epoch)+'th Epoch')
            print('w1 =  '+str(w1))
            print('w2 =  '+str(w2))
            break

    return w1,w2
    
```

```python
for i in range(len(y)):
    w1 = 0.12
    w2 = 0.12
    print((y[i] - (x1[i]*w1 + w2*x2[i]))*x1[i])
    print((y[i] - (x1[i]*w1 + w2*x2[i]))*x2[i])
    #print(activationFunction(x1[i],x2[i],w1,w2))
```

```python
w1, w2 = perceptronTraing(x1,x2,y)
```

```python
def lineQu(x1,w1,w2):
    pnt = []
    for i in range(len(x1)):
        x2 = -(w1/w2)*x1[i]+(0.5/w2)
        pnt.append(x2)
    return pnt
```

```python
x_2 = lineQu(x1,w1,w2)
```

```python
plt.plot(x1,x_2)
plt.title('OR Gate')
plt.scatter(x1,x2, c=y, cmap='viridis', s=10)
plt.show()
```
